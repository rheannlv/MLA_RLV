### **Renaming columns and getting statistics**

```python
pd.set_option('display.max_columns', None)
```

```python
df.rename(columns={'Date/Time':'date_time'}, inplace=True)
#Testing with one instance
```

```python
df.rename(columns={'Temp (C)':'temp_c', 'Dew Point Temp (C)':'dew_point', 'Rel Hum (%)':'relative_humidity', 'Wind Spd (km/h)':'wind_speed', 'Visibility (km)':'visibility', 'Stn Press (kPa)':'pressure', 'Weather':'weather'}, inplace=True)
#Renaming the remainder of the columns in bulk
```

```python
df.describe()
#Getting standard statistics of each column
```

### **Adding new temp_f column, converting to Fahrenheit and rearrange column**

```python
df['temp_f']=(df['temp_c'] * (9/5) + 32)
#Creating new column labelled temp_f and converting
```

```python
df = df[['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']]
#Rearranging columns
```

```python
df.head()
#Checking column name change
```

### **Display column data types, convert date_time column from object to datetime(ns)**

```python
df.dtypes
```

```python
df.date_time=pd.to_datetime(df.date_time)
#converting date_time
```

```python
df.date_time.dtype
#checking for successful conversion
```

### **Create monthly_avg as series**

```python
monthly_avg = df.groupby(df.date_time.dt.month)['temp_f'].mean()
```

```python
monthly_avg
```

```python
monthly_avg = pd.Series(monthly_avg)
```

### **Create three plots (scatter plot - temp vs relative humidity, line plot - monthly_avg temperature and box plot - entire dataframe)**

```python
df.plot(kind='scatter', x='temp_f', y='relative_humidity', title='Temperature vs Relative Humidity', color='green', figsize=(20,10))
plt.xlabel("Temperature (F)")
plt.ylabel("Relative Humidity")
```

```python
monthly_avg.plot(kind='line', x='date_time', y='monthly_avg', title='Average Temperature per Month', color='red', figsize=(20,10))
plt.xlabel("Month")
plt.ylabel("Temperature (F)")
```

```python
fig = plt.figure(figsize=(20,10))
sns.boxplot(data=df)
```
